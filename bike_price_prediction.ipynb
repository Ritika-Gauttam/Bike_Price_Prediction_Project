## Filtering

Find out bikes with TVS brand , owner = 1st owner , age <=2

```python
tvs_first_owner = tvs_brand[tvs_brand["owner"]=="First Owner"]
```

```python
tvs_age = tvs_first_owner[tvs_first_owner['age'] <= 2]
tvs_age
```

Find brand = " royal", km = >50Km, owner = 1st, age 2 sal se km  

```python
RoyalEnfiled = df[df['brand']== "Royal Enfield"]
RoyalEnfiled
```

```python
RoyalEnfiled_km = RoyalEnfiled[RoyalEnfiled["kms_driven"] >= 50]
RoyalEnfiled_km
```

```python
RoyalEnfiled_km_owner = RoyalEnfiled_km[RoyalEnfiled_km["owner"] == "First Owner"]
RoyalEnfiled_km_owner
```

```python
RoyalEnfiled_km_owner_age = RoyalEnfiled_km_owner[RoyalEnfiled_km_owner["age"] <= 2.0]
RoyalEnfiled_km_owner_age
```

```python
RoyalEnfiled_km_owner_age_price = RoyalEnfiled_km_owner_age[RoyalEnfiled_km_owner_age["price"] <= 150000.0]
RoyalEnfiled_km_owner_age_price
```

```python
sns.histplot(df['price'], kde=True)

plt.title('Distribution of Bike Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
```

```python
sns.scatterplot(x='kms_driven',y='price',data=df)
plt.title('Price vs Kilometers Driven')
plt.xlabel('Kilometers Driven')
plt.ylabel('Price')
```

```python
sns.barplot(x='owner',y='price',data=df,errorbar= None)
plt.title('Average Price by Owner Type')
plt.xlabel('Owner Type')
plt.ylabel('Average Price')
```

```python
sns.scatterplot(x='age',y='price',data=df)
plt.title('Price vs Bike Age')
plt.xlabel('Bike Age (Years)')
plt.ylabel('Price')
```

```python
sns.scatterplot(x='power',y='price',data=df)
plt.title('Price vs Engine Power')
plt.xlabel('Engine Power (cc)')
plt.ylabel('Price')
```

```python
brand_price = (df.groupby('brand')['price'].mean().sort_values())
brand_price.plot(kind='barh')
plt.title('Average Bike Price by Brand')
plt.xlabel('Average Price')
plt.ylabel('Brand')
```

```python
num_cols = ['price','kms_driven','age','power']
sns.heatmap(df[num_cols].corr(),annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
```
